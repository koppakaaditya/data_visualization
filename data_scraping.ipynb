```python
import requests
import pandas as pd
import time

API_KEY = "your_api_key"
# Full list of 250 stock symbols
symbols = [
    "AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "FB", "NVDA", "JPM", "JNJ", "V", 
    "WMT", "PG", "UNH", "DIS", "MA", "HD", "BAC", "PYPL", "VZ", "ADBE", 
    "NFLX", "CMCSA", "KO", "INTC", "PFE", "PEP", "T", "MRK", "XOM", "CSCO", 
    "ABT", "CRM", "ABBV", "MCD", "NKE", "WFC", "TMO", "LLY", "ORCL", "ACN", 
    "DHR", "COST", "CVX", "NEE", "NVS", "BMY", "SBUX", "MDT", "UNP", "HON", 
    "IBM", "RTX", "PM", "LIN", "QCOM", "MMM", "GILD", "BA", "UPS", "NOW", 
    "MS", "LOW", "INTU", "BKNG", "CAT", "TGT", "ISRG", "CB", "LMT", "AXP", 
    "SPGI", "PLD", "SYK", "EL", "MO", "MU", "USB", "AMAT", "DE", "BLK", 
    "SCHW", "AMD", "MDLZ", "ADI", "GS", "AMGN", "ZTS", "CSX", "C", "CI", 
    "ITW", "TFC", "FDX", "BDX", "ADP", "CL", "REGN", "WM", "VRTX", "ECL", 
    "NSC", "EW", "HUM", "ETN", "D", "GM", "LRCX", "F", "CME", "EMR", "ADI", 
    "AON", "LHX", "MRNA", "ICE", "COF", "ILMN", "MAR", "MCO", "KLAC", "SO", 
    "TRV", "STZ", "DG", "MPC", "PRU", "ORLY", "FTNT", "PGR", "BK", "SHW", 
    "SLB", "PH", "ADSK", "AEP", "LUV", "GPN", "IDXX", "PSA", "OXY", "APD", 
    "DD", "AFL", "ROST", "EBAY", "PPG", "EA", "SYY", "MTB", "TSCO", "PXD", 
    "VLO", "KHC", "MCK", "CTAS", "FAST", "TDG", "CARR", "BIIB", "WBA", 
    "HES", "KMB", "BKR", "A", "WMB", "ALL", "DOW", "OTIS", "CMG", "KEYS", 
    "CDW", "NEM", "ROK", "HLT", "WELL", "AVB", "HCA", "DLR", "EXC", "BXP", 
    "SPG", "KMI", "HPQ", "SRE", "MSCI", "ARE", "PEG", "AEE", "FIS", "PAYX", 
    "DTE", "PPL", "AIZ", "CNP", "NI", "AES", "VTR", "IRM", "ESS", "ETR", 
    "ED", "FE", "EVRG", "XEL", "PFG", "ZBH", "IQV", "WEC", "CMS", "NRG", 
    "AON", "TT", "MTD", "SIVB", "MKC", "VRSK", "CNC", "NDAQ", "VMC", 
    "XYL", "TSN", "ZION", "IFF", "LVS", "LEN", "WY", "WHR", "HSIC", "FFIV", 
    "EXPE", "NUE", "UAL", "TROW", "GWW", "DVA", "STT", "SJM", "LKQ", 
    "AIG", "NVR", "CBOE", "TRMB", "NRZ", "TPR", "ALB", "ZBRA", "ATO", 
    "UHS", "AES", "DGX", "DISH", "MGM", "HIG", "EXPD", "EQR", "CF", "TXT", 
    "PBCT", "HST", "HAS", "HBAN", "OKE", "LNC", "KMX", "NTRS", "CAG", 
    "FRC", "KEY", "RF", "DOV", "BWA", "WRB", "VFC", "AAP", "CLX", "BEN"
]
all_data = []

def fetch_technical_indicator(symbol, indicator, interval="daily", time_period=20):
    url = f"https://www.alphavantage.co/query?function={indicator}&symbol={symbol}&interval={interval}&time_period={time_period}&series_type=close&apikey={API_KEY}"
    response = requests.get(url)
    data = response.json()
    return data

for symbol in symbols:
    # Fetch daily time series data for the symbol
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={API_KEY}"
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"Error fetching data for {symbol}. Status code: {response.status_code}")
        continue
    
    data = response.json()

    # Fetch technical indicators for the symbol
    sma_data = fetch_technical_indicator(symbol, "SMA")  
    ema_data = fetch_technical_indicator(symbol, "EMA") 
    macd_data = fetch_technical_indicator(symbol, "MACD", interval="daily")  
    rsi_data = fetch_technical_indicator(symbol, "RSI") 
    bbands_data = fetch_technical_indicator(symbol, "BBANDS")  
    atr_data = fetch_technical_indicator(symbol, "ATR")  
    vwap_data = fetch_technical_indicator(symbol, "VWAP", interval="5min")  

    if 'Time Series (Daily)' in data:
        daily_data = data['Time Series (Daily)']
        
        for date, daily_values in daily_data.items():
            open_price = daily_values['1. open']
            high_price = daily_values['2. high']
            low_price = daily_values['3. low']
            close_price = daily_values['4. close']
            volume = daily_values['5. volume']

            sma = sma_data.get('Technical Analysis: SMA', {}).get(date, {}).get('SMA', None)
            ema = ema_data.get('Technical Analysis: EMA', {}).get(date, {}).get('EMA', None)
            macd = macd_data.get('Technical Analysis: MACD', {}).get(date, {}).get('MACD', None)
            rsi = rsi_data.get('Technical Analysis: RSI', {}).get(date, {}).get('RSI', None)
            bbands_upper = bbands_data.get('Technical Analysis: BBANDS', {}).get(date, {}).get('Real Upper Band', None)
            bbands_lower = bbands_data.get('Technical Analysis: BBANDS', {}).get(date, {}).get('Real Lower Band', None)
            atr = atr_data.get('Technical Analysis: ATR', {}).get(date, {}).get('ATR', None)
            vwap = vwap_data.get('Technical Analysis: VWAP', {}).get(date, {}).get('VWAP', None)

            all_data.append({
                'symbol': symbol,
                'date': date,
                'open': float(open_price),
                'high': float(high_price),
                'low': float(low_price),
                'close': float(close_price),
                'volume': int(volume),
                'SMA': float(sma) if sma else None,
                'EMA': float(ema) if ema else None,
                'MACD': float(macd) if macd else None,
                'RSI': float(rsi) if rsi else None,
                'BBANDS Upper': float(bbands_upper) if bbands_upper else None,
                'BBANDS Lower': float(bbands_lower) if bbands_lower else None,
                'ATR': float(atr) if atr else None,
                'VWAP': float(vwap) if vwap else None
            })
    else:
        print(f"No data found for {symbol}")
    
    # Sleep to respect the API rate limits (e.g., 5 API calls per minute)
    time.sleep(12)

df = pd.DataFrame(all_data)

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Filter the data to include only August and September
df_filtered = df[(df['date'] >= '2024-08-01') & (df['date'] <= '2024-08-01')]

# Sort the filtered data by date and symbol
df_filtered = df_filtered.sort_values(by=['date', 'symbol'])

print(df_filtered)

# Save the filtered data to a CSV file
# df_filtered.to_csv('stock_data_aug_sept.csv', index=False)
```
